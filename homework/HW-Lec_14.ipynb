```python
%matplotlib inline
# Brownian Motion
import numpy as np
from numpy import pi, sin, cos
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.animation import FuncAnimation
from IPython import display
```

```python

L1 = 101
L2 = 101
N = 5000

colors = mpl.colors.Colormap(name = "inferno", N=N)
c = np.linspace(0,1,N)

position = np.empty([N,2])
position[0,0] = int(L1/2)
position[0,1] = int(L2/2)
direction = np.array([[0,1],[0,-1],[1,0],[-1,0]])

np.random.seed(0)

# random walk
for i in range(1,N):
    position1 = [-1,-1]
    
    while min(position1)<0 or position1[0] >L1 or position1[1] >L2:
        move = np.random.randint(len(direction))
        position1 = position[i-1,:] + direction[move]
        
    position[i,:] = position1
```

```python
fig = plt.figure(figsize=[12, 12])

points = plt.plot([], "o")
point = points[0]
# other setup

plt.axis("scaled") # scales the plot so the output is square 
plt.xlim(0, L1)
plt.ylim(0, L2)
plt.title('Brownian motion')

def init_func():
    plt.axis("scaled") # scales the plot so the output is square 
    plt.xlim(0, L1)
    plt.ylim(0, L2)
    plt.title('Brownian motion')

def animate(frame):
    
    plt.scatter(position[frame,0], position[frame,1],cmap="hsv", norm=500 , s=10, alpha=0.5)
#     if frame ==0:
#             plt.text(position[0,0], position[0,1], 'start', fontsize=16)
#     if frame ==N-1:
#         plt.text(position[-1,0], position[-1,1], 'end', fontsize=16)
    


anim = FuncAnimation(fig, animate, frames=N, interval=30)
video = anim.to_html5_video()
html = display.HTML(video)
display.display(html)
plt.close()
    
    
```
