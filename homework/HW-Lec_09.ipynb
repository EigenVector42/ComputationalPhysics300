```python
data = np.loadtxt('http://www-personal.umich.edu/~mejn/cp/data/circular.txt',float)
#print(np.shape(data))

plt.imshow(data)
plt.title("circular.txt")
plt.show()
```

```python
data_hat = fft2(data)

# plt.plot(np.abs(data_hat))
# plt.show()

for i in range(len(data_hat)):
    for j in range(len(data_hat)):
        if np.abs(data_hat[i,j]) < 100:
            data_hat[i,j] = 0

# plt.plot(np.abs(data_hat))
# plt.show()
# plt.imshow((np.fft.fftshift(data_hat)).real)
# plt.show()
```

```python
restored_data=ifft2(data_hat)

plt.imshow(data)
plt.title("Original Image")
plt.savefig('circular.jpg')
plt.show()

plt.imshow(restored_data.real)

plt.title("Post Copression Image")
plt.savefig('circular-compressed.jpg')
plt.show()
```
