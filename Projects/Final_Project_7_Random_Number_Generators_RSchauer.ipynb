## 3.1 Psudeorandom Number Generators

### John von Neumann's Middle Square Method

>Any one who considers arithmetical methods of producing random digits is, of course, in a state of sin. For, as has been pointed out several times, there is no such thing as a random number-there are only methods to produce random numbers, and a strict arithmetic procedure of course is not such a method.
>
> <br>
> <b>John von Neumann</b>, from <i>"Various Techniques Used in Connection With Random Digits"</i>, 1951

John von Neumann proposed the first algorithm for a Random Number Generator in 1947 (von Neumann 1951). The algorithm is called The Middle Squares Method. It is a very simple algorithm. I have demonstrated it below:

<center>
<h3>Middle Square Method</h3><br><br>

<center>
<font size="3">
seed: 5769
$$\Downarrow$$
square the seed: $$5769^{2}$$
$$\Downarrow$$
Extract the "Middle" number:<br><br>
33<u><b>2813</b></u>61
$$\Downarrow$$
Add new number to list of Pseudo-Random Numbers:<br><br>
<b>2813</b><br>
$$\Downarrow$$
Square the new number:
$$2813^{2}$$
$$\Downarrow$$
Fill the front with zeros to ensure there are 8-digits to extract a middle number from:<br><br>
07<u><b>9129</b></u>69<br>
$$\Downarrow$$
Add new number to list of Pseudo-Random Numbers:<br><br>
<b>9129</b><br>
$$\Downarrow$$
Square the new number:
$$9129^{2}$$
$$\Downarrow$$
Rinse and repeat...<br><br>

</font>
</center>

There are several requirements for the algorithm to work. The n-digit starting value must be even, otherwise a unique "middle digit" might not be obtained. Also, the maximum period length of the sequence will not be longer that $8^{n}$ values ("Middle Square Method", Wikipedia 2020). Though this period limit is rarely reach as the sequence often converges to single value that will then repeat (often that value is zero). Below I have used the Python code provided on the Wikipedia page to demonstrate this issue with the algorithm. 

```python
"""
    Code copied from the The Middle Square Method Wikipedia page:

    https://en.wikipedia.org/wiki/Middle-square_method#Middle_Square_Weyl_Sequence_PRNG

"""

seed_number = int(input("Please enter a four digit number:\n[####] "))
number = seed_number
already_seen = set()
counter = 0

while number not in already_seen:
    counter += 1
    already_seen.add(number)
    """
        The section below is the Middle Square Algorithm.
        1. The input number is squared: number * number
        2. then converted into a string: str(number * number)
        3. The front digits are filled with zeries if necessary to
            bring the string up to 8-digit size: *.zfill(8)
        4. The string is sliced from the 2nd to the 5th indices to
            extract the "middle" number.
        5. Finally the string is converted into a 4-digit number:
            int(str(number * number).zfill(8)[2:6])
        
    """
    number = int(str(number * number).zfill(8)[2:6])  # zfill adds padding of zeroes
    print(f"#{counter}: {number}")

print(f"We began with {seed_number}, and"
      f" have repeated ourselves after {counter} steps"
      f" with {number}.")
```

## 3.3 Middle Square Weyl Sequence
### From terrible to amazing!

Von Neumann's Middle Square algorithm didn't have much going for it other than historical significance. That all changed in 2018 when Bernard Widynski modified the algorithm and created the Middle Square Weyl Sequence RNG (Widynski 2020). <b><u><i>This is possibly the fastest PRNG that passes all the standard statistical benchmark tests</i></u></b> ("Middle Square Method", Wikipedia 2020). The code for the algorithm is amazingly concise. Below is the C implementation for the algorithm:

<code>
    
    #include <stdint.h>

    uint64_t x = 0, w = 0

    // Must be odd (least significant bit is "1"), and upper 64-bits non-zero
    uint64_t s = 0xb5ad4eceda1ce2a9; // qualifying seed

    // return 32-bit number
    inline static uint32_t msws() {
        x *= x; // square the number
        w += s; // the weyl sequence
        x += w; // apply to x
        return x = (x>>32) | (x<<32); // return the middle 32-bits
    }
</code>
    
Source: [Aaron Toponce, Middle Square Weyl Sequence PRNG](https://pthree.org/2018/07/30/middle-square-weyl-sequence-prng/)
<br><br>
    
This code will generate a long period of well distributed random numbers.
<br><br>
Though the code is simple, the inner workings of the sequence to the algorithm are a bit more complicated.
    
The following explanation of the algorithm is taken from [Aaron Toponce, Middle Square Weyl Sequence PRNG](https://pthree.org/2018/07/30/middle-square-weyl-sequence-prng/)
<br><br>
    
***
Suppose "x = 0xace983fe671dbd09". Then "x" is a 64-bit number with the following bits:
> 
> 1010110011101001100000111111111001100111000111011011110100001001
***
When that number is squared, it becomes the 128-bit number 0x74ca9e5f63b6047f6a65456d9da04a51, or in binary:
> 01110100110010101001111001011111011000111011011000000100011111110110101001100101010001010110110110011101101000000100101001010001
***
But remember, "x" is a 64-bit number, so in our C code, only the bottom 64-bits are returned from that 128-bit number.

So "x" is really 0x6a65456d9da04a51, or in binary: 
> 0110101001100101010001010110110110011101101000000100101001010001
***
But the bits "01101010011001010100010101101101" are the 3rd 32-bits of the 128-bit number that was the result of squaring "x" (see above). They are the "middle" 32-bits that we're after. So, we're going to do something rather clever. We're going to swap the upper 32-bits with the lower, then return the lower 32-bits.

Effectively, what we're doing is "ABCD" -> "CDAB", then returning "AB". We do this via bit-shifting. 
***    
So, starting with:

> 0110101001100101010001010110110110011101101000000100101001010001
***
First, we bitshift the 64-bit number right 32-bits:

> 0110101001100101010001010110110110011101101000000100101001010001 >> 32
> = 0000000000000000000000000000000001101010011001010100010101101101
***
Then we bitshift "x" left 32-bits:

> 0110101001100101010001010110110110011101101000000100101001010001 << 32
> = 1001110110100000010010100101000100000000000000000000000000000000
***
Now we logically "or" them together:

> 0000000000000000000000000000000001101010011001010100010101101101 |
> 1001110110100000010010100101000100000000000000000000000000000000
***
 1001110110100000010010100101000101101010011001010100010101101101
*** 

<br><br>   
Our Original Seed was: 12459634972772121865 (0xace983fe671dbd09 in hexadecimal)
    
Our New Number is now: 11358159973766219117 (0x9da04a516a65456d in hexadecimal)
    
<br><br>
In May 2020, Widynski published a [paper](https://arxiv.org/pdf/1704.00358.pdf) with a further improvement of the algorithm, the Squares CBRNG, which is a Counters Based Random Number Generator.
    
<br><br>
The implementation of both the Middle Squares Weyl Sequence RNG and the Squares CBRNG are free from the creator and can be downloaded from this [location](https://mswsrng.wixsite.com/rand).

```python

def msws(seed=0xb5ad4eceda1ce2a9, 
         r_values=1, 
         scale=(2**32), 
         shifted=0):
    
    """
        Implementation of the Middle Squares Weyl Sequence
    """
    
    # C type casting to allow for bit shifting
    x = ctypes.c_uint64(0)
    w = ctypes.c_uint64(0)
    
    # list to hold the generated random numbers
    values = []

    # loop for generating the random number list
    for i in range(r_values):    
        
        # squaring the x value
        x.value *= x.value
        # adding the seed to the weyl sequence value
        w.value += seed
        # adding the new weyl sequence value to the square
        x.value += w.value
        
        # bit shifting to extract the "middle" value
        x.value = (x.value >> 32) | (x.value << 32)

        # convertingthe value to a 32-bit, this will extract the "middle" square
        x32 = ctypes.c_uint32(x.value).value

        # casting the number to align with the scale value
        if scale==1:
            x32 = int(x32/scale + shifted)
        elif scale==2**32:
            x32 = x32/scale + shifted
        else:
            tp = type(scale)
            x32 = tp(x32/scale + shifted) 
        
        # append the value to the list
        values.append(x32)
    
    return values    
```

```python
x_range = 1000

# s = 0xb5ad4eceda1ce2a9
s = seeds[int((10**7)*time()%(len(seeds)))]
# s = 212154654132134512

scale = 1
# scale = 2**32

timer_start = time()

int_values = msws(seed=s , r_values=x_range, scale=scale)

timer_end = time()
timer = timer_end - timer_start

print(f"Elapsed Time: {timer} seconds")
_x = np.linspace(0, x_range, x_range)
plt.scatter(_x, int_values)
plt.title(f"Scatter Plot of Random 32-bit Integers")
plt.show()

plt.hist(int_values, bins=11)
plt.title(f"Histogram of Distribution of Random 32-bit Integers")
plt.show()

print(min(int_values), max(int_values))
```
